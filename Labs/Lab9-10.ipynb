# Lab Part I: Runs Created

```python
# Runs Created = (hits + walks + hit by pitch) * TB / (at bats + walks + hit by pitch))
```

```python
# Mark McGwire 1998

# 70 HR, 147 R, 162 RBI, 583 AB, 509 TB, 0 SB, 0 CS, 0 HBP, 0 BB

m_runs_created = (152 + 162 + 6) * 383 / (509 + 162 + 6)

print(m_runs_created)
```

```python
# Sammy Sosa 1998
s_runs_created = (198 + 73 + 1) * 416 / (643 + 73 + 1)

print(s_runs_created)
```

# Lab Part II: Runs Created per Game

```python
# Runs Created per Game = Runs Created / .982(AB) - hits + GIDP + SF + SAC + CS / 26.72
```

```python
# Mark McGwire 1998

m_runs_created_per_game = 181 / (((.982 * 509) - 152 + 8 + 4 + 0 + 0) / 26.72)

print(m_runs_created_per_game)
```

```python
# Sammy Sosa 1998

s_runs_created_per_game = 158 / (((.982 * 643) - 198 + 20 + 5 + 0 + 9) / 26.72)

print(s_runs_created_per_game)
```
